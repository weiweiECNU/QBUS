```python
# Import Packages
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Read in csv files and plot top of training data
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
train.head()
```

```python
#Examine the missingness of the training data
train.isna().sum()
```

```python
# Examine missingness of the testing data
test.isna().sum()
```

```python
# Check different sub-classes from each variable. The pandas' crosstab feature allows us to plot the frequency of sub-groups
pd.crosstab(index=train["experiences_offered"], columns="count")
```

```python
# Clean training & testing data.
# Convert true/false to 1/0
train["instant_bookable"] = train.instant_bookable.map({'f':0, 't':1})
test["instant_bookable"] = test.instant_bookable.map({'f':0, 't':1})

# Deal with missing data by taking column mean
train["review_scores_cleanliness"].fillna(np.mean(train["review_scores_cleanliness"]), inplace=True)
test["review_scores_cleanliness"].fillna(np.mean(test["review_scores_cleanliness"]), inplace=True)

# Deal with missing data by filling NA with a 0 value
train["bathrooms"].fillna(0, inplace=True)
test["bathrooms"].fillna(0, inplace=True)    
```

```python
# print converted variable types
converted_variables = [["instant_bookable"], "review_scores_cleanliness", "bathrooms"]
for i in range(len(converted_variables)):
    print(train[converted_variables[i]])
```

```python
# Demonstrate frequency count for each category
pd.crosstab(index=train["property_type"], columns="count")
```

```python
# Encode smaller number of categorical variables for property type
train['property_type_encoded'] = np.where(train['property_type'].str.contains('House'), "House", (np.where(train['property_type'].str.contains('Apartment'), "Apartment", (np.where(train['property_type'].str.contains('Townhouse'), "Townhouse", "Other")))))
test['property_type_encoded'] = np.where(test['property_type'].str.contains('House'), "House", (np.where(test['property_type'].str.contains('Apartment'), "Apartment", (np.where(test['property_type'].str.contains('Townhouse'), "Townhouse", "Other")))))
```

```python
# Count frequency of different categories after sorting to smaller number of categories
pd.crosstab(index=train["property_type_encoded"], columns="count")
```

```python
# Split some of the clean training & testing data into 
indep_variables = ["accommodates", "beds", "bathrooms", "latitude", "longitude", "host_is_superhost", "review_scores_cleanliness", "bathrooms"]
dep_variable = ["price"]
```

```python
# Plot Exploratory Data Analysis
for i in range(len(indep_variables)):
    plt.scatter(train[indep_variables[i]], (train[dep_variable[0]]))
    plt.xlabel(str([indep_variables[i]]))
    plt.ylabel(str([dep_variable[0]]))
    plt.title("Training: Price vs "+str([indep_variables[i]]))
    plt.savefig("price vs"+indep_variables[i])
    plt.show()
```

```python
# Split some of the clean training & testing data into 
X_train = train[["accommodates", "latitude", "longitude"]]
y_train = train["price"]

# Create some test data
X_test = test[["accommodates", "latitude", "longitude"]]
```

```python
# Run the algorithm on a small number of variables & check results
ols = LinearRegression()
ols.fit(X_train, y_train)
y_pred = ols.predict(X_test)
print(y_pred)
```

```python
submission = pd.DataFrame(np.c_[test.index, y_pred], columns=['Id', 'price'])
submission['Id'] = submission['Id'].astype('int')
submission.to_csv('kaggle_submission.csv', index=False) 
print(submission)
```
