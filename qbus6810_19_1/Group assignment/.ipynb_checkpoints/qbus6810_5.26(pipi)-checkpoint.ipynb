```python
def convert_review_score(val):
    if val >= 90 and val <=100:
        return 'good'
    elif val >= 60 and val <= 89:
        return 'medium'
    elif val <= 59:
        return 'bad'

# Apply the function
train_data['review_scores_rating'] = train_data['review_scores_rating'].apply(convert_review_score)

# Count frequency for each category
pd.crosstab(index=train_data["review_scores_rating"], columns="count")
```

```python
def convert_review_score (val):
    if  val == 0:
        return 'No response'
    elif  val >= 8 and val <=10:
        return 'Good'
    elif val >= 5 and val <= 7:
        return 'Medium'
    elif val <= 4:
        return 'Poor'
    
# Apply the function
review_scores = ["review_scores_accuracy","review_scores_cleanliness","review_scores_checkin","review_scores_communication","review_scores_location","review_scores_value"]

for i in review_scores:
    train_data[i] = train_data[i].apply(convert_review_score)
    frequency = pd.crosstab(index=train_data[i], columns="count")
    print(frequency)
```

# Ridge Regression

```python


alphas = list(np.logspace(-15, 15, 151, base=2))

ridge = Pipeline((
    ('scaler', StandardScaler()),
    ('estimator', RidgeCV(alphas=alphas, cv=5)),
))

ridge.fit(X_train_log_scale, y_train)
```
