# 2.0 Feature Engineering

```python
# combine training + test set to do feature engineering together
# first drop column ID, as it is useless for predictions

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

test_ID = test['Id']

train.drop('Id', axis = 1, inplace = True)
test.drop('Id', axis = 1, inplace = True)

df_eng = train.append(test, ignore_index = True)
df_eng = df_eng.drop(columns = 'price')

# this is the way you could use to round decimal numbers
# check the descriptive statistics to check the correctness
np.round(df_eng.describe(),decimals = 3).T
```

# 2.1 Numerical Variables: Deal With the Skewness

```python
# log(p+1) transformation of highly skewed numeric features
numeric_feats = df_eng.dtypes[df_eng.dtypes != 'object'].index

# Check the skewness of all numerical features
skewed_feats = df_eng[numeric_feats].apply(
    lambda x:skew(x.dropna())).sort_values(ascending=False)

# Skew in numerical features
skewness = pd.DataFrame({'Skew' :skewed_feats})
skewness.head(20)
```

```python
# transform data if skewness > 1.5
skewness = skewness[skewness['Skew'] > 1.5]

skewed_features = skewness.index

for feat in skewed_features:
    df_eng[feat] = np.log1p(df_eng[feat])

#comment:
# except lat, you can't transfer it as it's value is negative
# you can transfer it's abs value.or just leave it
```

# 2.2 Missing Value

```python
# type one: review-irralevant
# bathroom, security_deposit and cleaning_fee

# 1.bathroom: check the relationship between bathrom and bedroom
plt.scatter(x = df_eng['bathrooms'], y = df_eng['bedrooms'], s = 10)
plt.title('Correlation Between Bathrooms and Bedrooms')
plt.ylabel('Bedrooms')
plt.xlabel('Bathrooms')
plt.show()

# use bedrooms to fill bathrooms
df_eng['bathrooms'] = df_eng['bathrooms'].fillna(np.log1p(df_eng['bedrooms']))
```

```python
# 2. security_deposit and cleaning_fee: fillna(0)
df_eng['security_deposit'] = df_eng['security_deposit'].fillna(0)
df_eng['cleaning_fee'] = df_eng['cleaning_fee'].fillna(0)
```

```python
# type two: review realated
# use 'NO Reviews' for the 7 scores
# use 0 for 'per month'

df_eng['reviews_per_month'] = df_eng['reviews_per_month'].fillna(0)
df_eng = df_eng.fillna('No Reviews')
```

```python
# check: any more null values left
for col in df_eng:
    if(df_eng[col].isnull().any()):
        print(col)
```

# 2.3  Numerical Variables Transfer to Categorical

```python

def convert_scores_buckets(val):
    if val == 'No Reviews':
        return 'No Reviews'
    elif val >= 90.0:
        return 'excellent'
    elif val >= 80.0 and val < 90.0:
        return 'good'
    elif val >= 70.0 and val < 80.0:
        return 'not bad'
    elif val >= 60.0 and val < 70.0:
        return 'not too bad'
    elif val < 60.0:
        return 'poor'
    

def convert_scores_buckets_small(val):
    if val == 'No Reviews':
        return 'No Reviews'
    elif val >= 9.0:
        return 'excellent'
    elif val >= 8.0 and val < 9.0:
        return 'good'
    elif val >= 7.0 and val < 8.0:
        return 'not bad'
    elif val >= 6.0 and val < 7.0:
        return 'not too bad'
    elif val < 6.0:
        return 'poor'
```

```python
# transfer to categorical
df_eng['review_scores_accuracy'] = df_eng['review_scores_accuracy'].apply(convert_scores_buckets_small)
df_eng['review_scores_cleanliness'] = df_eng['review_scores_cleanliness'].apply(convert_scores_buckets_small)

df_eng['review_scores_checkin'] = df_eng['review_scores_checkin'].apply(convert_scores_buckets_small)

df_eng['review_scores_communication'] = df_eng['review_scores_communication'].apply(convert_scores_buckets_small)

df_eng['review_scores_location'] = df_eng['review_scores_location'].apply(convert_scores_buckets_small)

df_eng['review_scores_value'] = df_eng['review_scores_value'].apply(convert_scores_buckets_small)

df_eng['review_scores_rating'] = df_eng['review_scores_rating'].apply(convert_scores_buckets)

```

# 3.0 Data Modelling

```python
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import make_pipeline
from sklearn import ensemble
from sklearn.cross_validation import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression, RidgeCV, LassoCV, ElasticNetCV
from sklearn.ensemble import GradientBoostingRegressor
```

```python
X_train = df_eng.iloc[:383, :-1]
X_test= df_eng.iloc[383:,:-1]
y_train = y_train_log
#note: validation set could potentially be more important than public LB
X, X_val,y,y_val=train_test_split(X_train,y_train,test_size=0.2,random_state=0)
```

```python
def rmse(y,y_pred):
    return np.sqrt(mean_squared_error(y,y_pred))
```

# 3.1 OLS

```python
ols=LinearRegression()
ols.fit(X,y)
y_val_ols=ols.predict(X_val)
print('MLR Log_RMSE on Validation_set_1:',rmse(y_val,y_val_ols))
print('MLR RMSE on Validation_set_1:',rmse(np.expm1(y_val),np.expm1(y_val_ols)))

#fit with 383 training set
ols.fit(X_train,y_train)
y_train_ols=ols.predict(X_train)
y_test_ols=np.expm1(ols.predict(X_test))
print('MLR Log_RMSE on training_set_1:',rmse(y_train,y_train_ols))
print('MLR RMSE on training_set_1:',rmse(np.expm1(y_train),np.expm1(y_train_ols)))
```

```python
np.mean(np.exp(y_train - y_train_ols))
```

```python
y_pred = y_test_ols * np.mean(np.exp(y_train - y_train_ols))
```

```python
# check the predicted price performance on validation_set_1
# log_tranformed y
plt.figure()
plt.scatter(y_val_ols, y_val, s=10)
plt.title('MLR_prediction(log)')
plt.xlabel('Predicted Price')
plt.ylabel('Actual Price')
plt.plot([min(y_val), max(y_val)],[min(y_val), max(y_val)],color = 'red')
plt.show()                            

# original y
plt.figure()
plt.scatter(np.expm1(y_val_ols), np.expm1(y_val), s=10)
plt.title('MLR_prediction')
plt.xlabel('Predicted Price')
plt.ylabel('Actual Price')
plt.plot([min(np.expm1(y_val)), max(np.expm1(y_val))],[min(np.expm1(y_val)), max(np.expm1(y_val))],color = 'red')
plt.show()
```
