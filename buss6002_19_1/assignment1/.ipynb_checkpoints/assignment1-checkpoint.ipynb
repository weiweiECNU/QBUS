```python
# 方法2
import statsmodels.api as sm

from statsmodels.tools.eval_measures import rmse


X = a[['Gr Liv Area',"Total Bsmt SF"]]
y = a["SalePrice"]


X_with_intercept = sm.add_constant(X)

X_with_intercept.head()


```

```python
model1  = sm.OLS(y,X)
results = model1.fit()

results.params
```

```python
results.summary()
```

```python
y_pr1 = results.params[0] * a['Gr Liv Area'] + results.params[1]* a['Total Bsmt SF']
```

```python
rmse(y,y_pr1)
```

```python
model2  = sm.OLS(y,X_with_intercept)
results2 = model2.fit()

results2.params
```

```python
results2.summary()
```

```python
results2.params[2]
```

```python
y_pr2 =  results2.params[1] * a['Gr Liv Area'] + results2.params[2]* a['Total Bsmt SF'] + results2.params[0]
```

```python
rmse(y,y_pr2)
```

```python
def search_corr( dummy_fields ):
    
    h1 = house["SalePrice"]

    for each in dummy_fields:
        dummies = pd.get_dummies( house.loc[:, each], prefix=each ) 
        h1 = pd.concat( [h1, dummies], axis = 1 )
    
    return h1.corr()["SalePrice"].sort_values(ascending=False)
```

```python
X = house[['Gr Liv Area',"Garage Cars","Garage Area","1st Flr SF","Year Built","Total Bsmt SF"]]

y = house["SalePrice"]

X_with_intercept = sm.add_constant(X)

X_with_intercept.head()

model3  = sm.OLS(y,X_with_intercept)
results3 = model3.fit()
```

```python
results3.summary()

```

```python
y_pr3 = results3.params[0] + results3.params[1] * house["Gr Liv Area"] + results3.params[2] * house["Garage Cars"] + results3.params[3] * house["Garage Area"] \
        + results3.params[4] * house["1st Flr SF"] + results3.params[5] * house["Year Built"] + results3.params[6] * house["Total Bsmt SF"]

rmse(y_pr3,y)
```
