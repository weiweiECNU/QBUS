# PCA 

```python
# 转成 矩阵
X = tfidf.todense()

# 降维 50维
from sklearn.decomposition import PCA
pca = PCA(n_components=50)
pca.fit(X)
```

```python
# PCA 维度分析图 TUT7
cum_var = np.cumsum(pca.explained_variance_ratio_) * 100

plt.plot(np.arange(1, 51), cum_var, marker = "o")
plt.xlabel("Number of Components")
plt.ylabel("Variance")
plt.title("Scree Plot (Number of Components vs Variance)")
plt.show()
```

```python

# 对 test set 做 tfidf 和 pca 

from sklearn.feature_extraction.text import TfidfVectorizer
corpus = pred['REVIEW_TEXT']
clf = TfidfVectorizer()
tfidf = clf.fit_transform(corpus)
X = tfidf.todense()
pca = PCA(n_components=50)
b = pd.DataFrame(pca.fit_transform(X))
```

```python
# 选择作为分类依据的 变量
variable = ["RATING","VERIFIED_PURCHASE","PRODUCT_CATEGORY","PRODUCT_TITLE","PRODUCT_ID","TextLength","TitleLength","Textcap","Titlecap"] +[i for i in np.arange(50)]
variable
```

```python
# 分 train 和 validation set 

from sklearn.model_selection import train_test_split

X = df2[variable]
y = df2["LABEL"]

X_train, X_validation, y_train, y_validation = train_test_split(X, y,test_size=0.2,random_state=0)
```

```python
# 忽略警告
import warnings
warnings.filterwarnings("ignore")
```

```python
# gridsearchcv 调 logistic 的 参数
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

lr = LogisticRegression()
grid_value = {'solver':['sag','liblinear','lbfgs'],
             'C':[1, 2,5,10,20,50]}
model_lr = GridSearchCV(lr, cv=5, param_grid=grid_value)

lr_cv = model_lr.fit(X_train, y_train).best_estimator_
```

```python
# 预测 validation  
y_pred = lr_cv.predict(X_validation)
```

```python
# validation set 的 预测分析
from sklearn.metrics import classification_report

print(classification_report(y_validation, y_pred))
```
