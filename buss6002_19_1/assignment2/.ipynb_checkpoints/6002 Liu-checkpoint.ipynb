```python
def clean(df):
    df["RATING"].fillna( df["RATING"].mean(),inplace = True )
    df["VERIFIED_PURCHASE"].fillna("N", inplace = True )
    df["PRODUCT_CATEGORY"].fillna("None", inplace = True )
    df["PRODUCT_TITLE"].fillna(" ", inplace = True )
    df["REVIEW_TITLE"].fillna(" ", inplace = True )
    df["REVIEW_TEXT"].fillna(" ", inplace = True )
    df["REVIEW"] = df['REVIEW_TITLE'].str.cat(df['REVIEW_TEXT'], sep = ". ")
    df["VERIFIED_PURCHASE"] = df.VERIFIED_PURCHASE.map({'N':0, 'Y':1})
```

# Feature Engineering:

## Extract useful features

### length of review

```python
import nltk
nltk.download('punkt')
from nltk.stem.snowball import EnglishStemmer
snowball_stemmer = EnglishStemmer()
```

```python
def stem_tweet(tweet):
    """Break up a sentence, stem each word, and put it back together."""
    words = nltk.word_tokenize(tweet)
    stemmed_tweet = ' '.join([snowball_stemmer.stem(word) for word in words])
    return stemmed_tweet 

def get_length(tweet):
    """Get the length of each review. """
    words = nltk.word_tokenize(tweet)
    return len(words)
```

```python
#sum(map(str.isupper, train.REVIEW[4].split())) 
```

### percentage of capitalized words



```python
def percent_capitalized(tweet):
    """Count the number of the capitalized words in each review. """
    return sum(map(str.isupper, tweet.split())) / len(nltk.word_tokenize(tweet))
```

```python
#reviews['PRODUCT_TITLE'].str.count(r'[A-Z]')/reviews['PRODUCT_TITLE'].str.len()
```

```python
def Extract_features(df):
    df['stemmed_REVIEW'] = df.REVIEW.apply(stem_tweet)
    df["REVIEW_length"] = df.REVIEW.apply(get_length)
    df['REVIEW_pcw'] = df.REVIEW.apply(percent_capitalized)
    #df["PRODUCT_length"] = df.REVIEW.apply(get_length)
    return df
```

## TFIDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

def get_tfidf(df):
    """ calculate the tfidf of the review """
    corpus = df['stemmed_REVIEW']
    clf = TfidfVectorizer(stop_words = 'english',max_features=2000,min_df=2,max_df=0.9  )
    tfidf = clf.fit_transform(corpus)
    return tfidf.todense()
```

## PCA

```python
from sklearn.decomposition import PCA
```

```python
def visualise_variance(pca):
    """Visualise the vatiance of each component."""
    cum_var = np.cumsum(pca.explained_variance_ratio_) * 100
    plt.figure(figsize=(16,9))
    plt.plot(np.arange(1, pca.n_components_+1), cum_var, marker = "o")
    plt.xlabel("Number of Components")
    plt.ylabel("Variance")
    plt.title("Scree Plot (Number of Components vs Variance)")
    plt.show()
    
```

```python
def add_tfidf(df, components):
    """ calculate the tfidf and then pca the result. """
    tfidf = get_tfidf(df)
    pca = PCA(n_components=components)
    pca.fit(tfidf)
    
    visualise_variance(pca)
    
    tfidf_df = pd.DataFrame(pca.fit_transform(tfidf))
    
    return pd.concat([df,tfidf_df],axis = 1)
```

```python
train = add_tfidf(train ,100)
```

```python
test = add_tfidf(test ,100)
```

# Advance model

## Analysis the sentiment

```python
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sent = SentimentIntensityAnalyzer()

def sentiment_analysis(tweets):
    return sent.polarity_scores(tweets)['compound']
```

## Count the number of punctuation

```python
import string
def count_punctuation(s):
    count = lambda l1, l2: len(list(filter(lambda c: c in l2, l1)))
    return count(s,set(string.punctuation)) 
```

```python
def Extract_advance_features(df):
    df['sentiment'] = df.stemmed_REVIEW.apply(sentiment_analysis)
    df["num_punctuation"] = df.stemmed_REVIEW.apply(count_punctuation)
```

```python
Extract_advance_features(train)
Extract_advance_features(test)
```

```python
features = ["RATING","VERIFIED_PURCHASE","REVIEW_length","REVIEW_pcw","sentiment","num_punctuation"]
variable = features + [i for i in np.arange(100)]
X = train[variable]
y = train["LABEL"]
X_train, X_validation, y_train, y_validation = train_test_split(X, y,test_size=0.2,random_state=0)
```
