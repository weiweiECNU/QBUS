```python
house = pd.read_sql("SELECT * FROM houses;", conn)

house.head()
```

```python
a = house[['Gr Liv Area',"Total Bsmt SF","Year Built","Sale Condition",'SalePrice']]

a = a[ a["Sale Condition"] == "Normal"]


a.dropna(inplace = True)


a.iloc[:,2] = -(a.iloc[:,2] - 2010)


a.corr()


# GLA TBSF
```

```python
a
```

```python
from sklearn.linear_model import LinearRegression

X = a[['Gr Liv Area',"Total Bsmt SF"]]
y = a["SalePrice"]

reg1 = LinearRegression()

reg1.fit(X, y)
```

```python
reg1.coef_
```

```python
reg1.intercept_
```

```python
reg2 = LinearRegression(fit_intercept = False)

reg2.fit(X, y)
```

```python
reg2.coef_
```

```python
reg2.intercept_
```
