### 1. scale

所有的X和y，都要除以他们同一列上的最大值。

example：

```python
X = np.array([[4,7],[1,8],[3,2]])
y = np.array([5,2,3]).reshape(-1,1) #注意这里！！！！一定要把y变成N*1的形式！！！！
print(X)
print(y)
```

```python
X_max = np.max(X, axis=0)
print(X_max)

X_scaled = X/X_max
print(X_scaled)
```

```python
y_max = np.max(y)
print(y_max)

y_scaled = y/y_max
print(y_scaled)
```

### 2. forward propogation

上面的例子是:

Input layer size: 2

Hidden layer size: 3

Output layer size: 1

#### 2.1 Step 1：计算Z2

Input layer: 3行2列，也就是我们输入的X有3条数据，每条数据有2个feature。 这里(3\*2), input layer size: 2。

Hidden layer: 这里的size为3，这里的3指的是我们的hidden layer有3个unit，（甚至可以理解为这个时候的数据变成了3列），原来数据(3\*2)到这里变成了X(3\*2)乘W1(2\*3)，也就是 Z2=（3\*3）的矩阵。注意这里算出来的只是Z2，还需要通过activate function，才是这一层最终的结果。

```python
print(X_scaled)

np.random.seed(0)
input_layer_size = 2
hidden_layer_size = 3
W1= np.random.randn(input_layer_size, hidden_layer_size)  
print(W1)
#上面这几行只是随机生成一个W1,老师给的代码已经有这部分了，在做assignment的时候不需要再抄一遍了！

Z2 = X_scaled.dot(W1)
print(Z2)
```

#### Step 2: 计算a2


a2 = activate(Z2)

activation function有很多种，这题要求使用的是sigmod function:$$\sigma(z)= \frac{1}{1+e^{-z}}$$

```python
a2 = 1/(1+np.exp(-Z2))
print(a2)
```

#### Step 3&4: 计算Z2和a3


```python
np.random.seed(0)
hidden_layer_size = 3
output_layer_size= 1
W2= np.random.randn(hidden_layer_size, output_layer_size)  
print(W1)
#上面这几行只是随机生成一个W2,老师给的代码已经有这部分了，在做assignment的时候不需要再抄一遍了！

Z3 = a2.dot(W2)
print(Z3)

a3 = 1/(1+np.exp(-Z3))
print(a3)
```

### 3.计算loss
$$L(\boldsymbol{W})=\frac{1}{2}\sum_{n=1}^N(f(x_n,\boldsymbol{W})-t_n)^2$$

f(x,W)就是预测值,也就是a3，t是实际值，也就是y_scaled

```python
loss = np.sum((a3-y_scaled)**2)/2
print(loss)
```

### 4. Backpropogation

我们的目的是更新W1和W2，所以需要分别计算L对于W1和W2的梯度，公式推导这里就不详细讲了，期末复习时候会带着大家推导的。

```python
sigmod_d_Z3 = (np.exp(-Z3))/((1+np.exp(-Z3))**2)

#或者按照下面这么写，因为sigmod_Z3其实就是a3
sigma_d_Z3_2 = np.multiply(a3,(1-a3))

print(sigmod_d_Z3)
print(sigma_d_Z3_2)
```

计算sigma3

```python
print((a3-y_scaled).shape)
print(sigmod_d_Z3.shape)

sigma3 = np.multiply(a3-y_scaled, sigmod_d_Z3)
print(sigma3)
```

```python
W2_gradient = a2.T.dot(sigma3)
print(W2_gradient)
```

```python
sigma2= np.multiply(sigma3.dot(W2.T), (np.exp(-Z3))/((1+np.exp(-Z3))**2))
W1_gradient= X.T.dot(sigma2)
```

update W2和W1  **这一部分必须是算出W1的梯度之后再算！！！！！！** 因为在计算W1的梯度的时候，我们仍旧需要上一轮的W2

```python
alpha= 0.0001  #注意这里是老师已经给出的代码部分，不要复制这里
W2 = W2- alpha*W2_gradient  
print(W2)

W1 = W1- alpha*W1_gradient  #自己写这部分的时候注意，不要重复运行这个part好多次，每运行一次W1和W2都会变一次。
print(W1)
```

#### keras

如果用keras做，记得先用scaler去把X和y变成0到1之间的数。

本地用CPU跑keras可能会比较慢，如果想配置GPU版本的tensorflow或者theano，可能会把自己的python搞崩溃，大家谨慎做。

```python
from sklearn.preprocessing import MinMaxScaler
scaler_x = MinMaxScaler(feature_range=(0, 1))
scaler_y = MinMaxScaler(feature_range=(0, 1))

X_new_train = scaler_x.fit_transform(X_train)
y_new_train = scaler_y.fit_transform(y_train.reshape(-1,1))
```

```python
from keras.layers.core import Dense
from keras.models import Sequential


# Define the network architecture
model = Sequential()
n_features = X_new_train.shape[1]

# First layer, input size is the number of input features and output size is 1
# We are doing a linear regression, and let's use linearactivation
model.add(Dense(64, input_dim=n_features, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(1, activation='linear'))
#注意这里的层数，包括每层的unit数，都是人为定的，大家可以随便定，我只是随手写了几层。


# Use mean squared error for the loss metric and use the ADAM backprop algorithm
model.compile(loss='mean_squared_error', optimizer='adam')

# Train the network (learn the weights)
# We need to convert from DataFrame to NumpyArray
history = model.fit(X_new_train, y_new_train, epochs=5, batch_size=1000, verbose=2, validation_split=0.2)

#epoch代表了训练的轮数，轮数越多越准确，但是可能会出现overfit，所以我们取validation_split=0.2
#这样就用Cross validation解决overfit的问题了

#这里我为了节省时间就写了5个epoch
```

记得在这里预测X_test的时候，要先把X_test用scaler_x去transform，然后预测出来的Y，再用scaler_y去inverse_transform，这样就能得到和原始的y_train一样的格式了。

```python
y_pred = scaler_y.inverse_transform(model.predict(scaler_x.transform(X_test)))

print(np.sqrt(mean_squared_error(np.exp(y_pred), np.exp(y_test)))) 
#因为这里演示的是y取了log之后的，所以这里依旧要np.exp
```
